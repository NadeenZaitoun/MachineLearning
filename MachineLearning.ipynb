## Data Pre-processing

```python
df.isnull().head()
```

```python
df.drop(['event_time', 'price' , 'user_id' , 'user_session' ], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df['category_code'] = df['category_code'].replace(np.nan, 'Unknown')
```

```python
df['brand'] = df['brand'].replace(np.nan, 'Unknown')
```

```python
df
```

```python
df.drop(df[(df['event_type'] == 'cart')].index, inplace=True)
```

```python
from numpy.ma.core import count
data = df.groupby('product_id').agg(number_views = pd.NamedAgg('event_type', lambda x: (x == 'view').sum()), number_purchase = pd.NamedAgg('event_type', lambda y: (y == 'purchase').sum()))
data
```
